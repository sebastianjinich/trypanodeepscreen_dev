```python
from utils import get_actives_inactives_similarity_flat, get_braunblanquet_ASP_triangle
import pandas as pd


targets = dict()

for tarnum in (221,262,2581,2850,4072,4657,5567):

    target_path = f"/home/sjinich/disco/TrypanoDEEPscreen/data/processed/CHEMBL{tarnum}_chemblv34.csv"
    target = target_path.split("/")[-1].split("_")[0]
    df_target = pd.read_csv(target_path)
    similarity_triangle = get_braunblanquet_ASP_triangle(df_target)
    actives, inactives = get_actives_inactives_similarity_flat(df_target, similarity_triangle)
    inactives =inactives.dropna().reset_index(drop=True)
    actives = actives.dropna().reset_index(drop=True)
    targets[f"CHEMBL{tarnum}"] = (actives,inactives)
```

```python
for key in targets.keys():
    act, inact = targets[key]
    act.to_pickle(f"/home/sjinich/disco/TrypanoDEEPscreen/data_preprocessing/data_plots_ASP_braunblanquet/similarities/{key}_active.pkl")
    inact.to_pickle(f"/home/sjinich/disco/TrypanoDEEPscreen/data_preprocessing/data_plots_ASP_braunblanquet/similarities/{key}_inactive.pkl")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_histogram_with_two_y_axes(actives, inactives, target_name: str, outputpath, width=8, height=6, inset=True, inset_pos=[0.45, 0.4, 0.35, 0.25], inset_xrange=[0.6,1], inset_yrange_a=[0,200],inset_yrange_i=[0,1000],inset_bins=20):
    title = f"Similitud entre activos e inactivos de {target_name}"

    # Crear la figura y los dos ejes principales
    fig, ax1 = plt.subplots(figsize=(width, height))
    ax2 = ax1.twinx()  # Segundo eje Y

    # Colores para los histogramas
    color_activos = "blue"
    color_inactivos = "#40c365"
    alpha = 0.5
    edge_color = "black"

    # Graficar el histograma para los activos
    sns.histplot(
        actives["Similarity"],
        bins=50,
        ax=ax1,
        color=color_activos,
        kde=False,
        alpha=alpha,
        edgecolor=edge_color,
        label="Activos"
    )
    ax1.set_ylabel("Frecuencia (Activos)", color=color_activos)
    ax1.tick_params(axis='y', labelcolor=color_activos)

    # Graficar el histograma para los inactivos
    sns.histplot(
        inactives["Similarity"],
        bins=50,
        ax=ax2,
        color=color_inactivos,
        kde=False,
        alpha=alpha,
        edgecolor=edge_color,
        label="Inactivos"
    )
    ax2.set_ylabel("Frecuencia (Inactivos)", color=color_inactivos)
    ax2.tick_params(axis='y', labelcolor=color_inactivos)

    # Configurar el título y las etiquetas
    ax1.set_xlabel("Similitud química")
    plt.title(title)

    # Crear la leyenda combinada
    handles_1, labels_1 = ax1.get_legend_handles_labels()
    handles_2, labels_2 = ax2.get_legend_handles_labels()
    ax1.legend(handles=handles_1 + handles_2, labels=labels_1 + labels_2, loc="upper right")

    if inset:
        # Crear inset con dos ejes separados
        inset_ax1 = fig.add_axes(inset_pos)  # Eje para los activos
        inset_ax2 = inset_ax1.twinx()  # Eje para los inactivos

        # Graficar el histograma para los activos en el inset
        sns.histplot(
            actives["Similarity"],
            bins=inset_bins,
            ax=inset_ax1,
            color=color_activos,
            kde=False,
            alpha=alpha,
            edgecolor=edge_color,
            label="Activos"
        )
        inset_ax1.set_xlim(inset_xrange[0],inset_xrange[1])  # Rango limitado del inset
        inset_ax1.set_ylim(inset_yrange_a[0],inset_yrange_a[1])
        inset_ax1.set_xlabel("Similitud química")
        inset_ax1.set_ylabel("Frecuencia (Activos)", color=color_activos, fontsize=8)
        inset_ax1.tick_params(axis='y', labelcolor=color_activos, labelsize=8)
        inset_ax1.tick_params(axis='x', labelsize=8)


        # Graficar el histograma para los inactivos en el inset
        sns.histplot(
            inactives["Similarity"],
            bins=inset_bins,
            ax=inset_ax2,
            color=color_inactivos,
            kde=False,
            alpha=alpha,
            edgecolor=edge_color,
            label="Inactivos"
        )
        inset_ax2.set_xlim(inset_xrange[0],inset_xrange[1])
        inset_ax2.set_ylim(inset_yrange_i[0],inset_yrange_i[1])  # Rango limitado del inset
        inset_ax2.set_ylabel("Frecuencia (Inactivos)", color=color_inactivos, fontsize=8)
        inset_ax2.tick_params(axis='y', labelcolor=color_inactivos, labelsize=8)

        # Configurar título y etiquetas del inset
        inset_ax1.set_title("Detalle (Similitud alta)", fontsize=10)

    plt.savefig(outputpath, dpi=300, bbox_inches="tight")
 
    # Mostrar el gráfico
    plt.show()
```
